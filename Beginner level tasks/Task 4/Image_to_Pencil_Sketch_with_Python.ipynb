First , we import all the required libraries :

```python
import cv2
import io # io Library used for various input and ouput operations
from google.colab import files # this library used for dealing with files
from PIL import Image
from IPython.display import display
import matplotlib.pyplot as plt
%matplotlib inline
```

Uploading the image .

```python
img=cv2.imread('/content/tj.jpg')
```

Now , we resize the image according to our requirements.

```python
img = cv2.resize(img, (600, 500))
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

Now , we convert the uploaded image into grayscale.

```python
new_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(new_image))
```

Now , we invert the image .

```python
invert = cv2.bitwise_not(new_image)
display(Image.fromarray(invert))
```

Now , we have to smoothen the image.<br>
That is , we will be removing noise from the image .

```python
blur_image = cv2.GaussianBlur(invert, (5, 5), 0, 0) #GaussianBlur is is used fotr removing the noise 
display(Image.fromarray(blur_image))
```

Now , we will revert the image back to its original form

```python
invert_blur_image = 255 - blur_image
pencil_sketch = cv2.divide(new_image, invert_blur_image, scale=250.0)
display(Image.fromarray(pencil_sketch))
```

displaying the pencil sketch version of the uploaded image and originally uploaded image.

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
display(Image.fromarray(pencil_sketch))
```
